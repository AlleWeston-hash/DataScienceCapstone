```python
import requests
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
response = requests.get(URL)

if response.status_code == 200:
    data = io.StringIO(response.text)
    import pandas as pd
    spacex_df = pd.read_csv(data)
    print(spacex_df.head())
else:
    print(f"Failed to fetch data. Status code: {response.status_code}")
```

```python

def classify_landing_outcome(outcome):
    if 'Success' in outcome:
        return 'green' 
    elif 'Failure' in outcome:
        return 'red' 
    else:
        return 'orange' 


merged_df = launch_sites_df.merge(spacex_df[['Launch Site', 'Landing Outcome']], on='Launch Site', how='left')
for index, row in merged_df.iterrows():
    launch_site = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    landing_outcome = row['Landing Outcome']
    color = classify_landing_outcome(landing_outcome)
    

    circle = folium.Circle(
        location=[lat, long],
        radius=1000,  
        color=color,
        fill=True
    ).add_child(folium.Popup(f"{launch_site}: {landing_outcome}"))
    
    marker = folium.map.Marker(
        location=[lat, long],
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html='<div style="font-size: 12; color:%s;"><b>%s</b></div>' % (color, launch_site)
        )
    )
    
    site_map.add_child(circle)
    site_map.add_child(marker)

site_map
```

```python

from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster()

def get_marker_color(landing_class):
    if landing_class == 1:
        return 'green'
    elif landing_class == 0:
        return 'red'
    else:
        return 'blue'

for index, row in spacex_df.iterrows():
    launch_site = row['Launch Site']
    lat = row['Lat']
    long = row['Long']
    landing_class = row['class']  
    

    marker_color = get_marker_color(landing_class)
    
    marker = folium.CircleMarker(
        location=[lat, long],
        radius=8,
        color=marker_color,
        fill=True,
        fill_color=marker_color,
        fill_opacity=0.7
    )
    
    marker_cluster.add_child(marker)

site_map.add_child(marker_cluster)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    R = 7773.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python

coastline_lat = 28.56367
coastline_lon = -80.57163
launch_site_lat = 28.5721
launch_site_lon = -80.6480  
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print(f"Distance from the launch site to the closest coastline: {distance_coastline:.2f} km")
```
